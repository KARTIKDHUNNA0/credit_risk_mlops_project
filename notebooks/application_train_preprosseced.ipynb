1. EXT_SOURCE_3 (17.7% missing) — IMPORTANT FEATURE → IMPUTE


```python
mtraindf['EXT_SOURCE_3'] = mtraindf['EXT_SOURCE_3'].fillna(mtraindf['EXT_SOURCE_3'].median())
```

2. The AMT_REQ_CREDIT_BUREAU_ family (12% missing)*

```python
bureau_cols = [
    'AMT_REQ_CREDIT_BUREAU_WEEK','AMT_REQ_CREDIT_BUREAU_QRT',
    'AMT_REQ_CREDIT_BUREAU_MON','AMT_REQ_CREDIT_BUREAU_YEAR',
    'AMT_REQ_CREDIT_BUREAU_DAY','AMT_REQ_CREDIT_BUREAU_HOUR'
]

for col in bureau_cols:
    mtraindf[col] = mtraindf[col].fillna(0)
```

3. Social circle features 0.000595 missing  is very less

```python
soc_cols = [
    'OBS_30_CNT_SOCIAL_CIRCLE','DEF_30_CNT_SOCIAL_CIRCLE',
    'OBS_60_CNT_SOCIAL_CIRCLE','DEF_60_CNT_SOCIAL_CIRCLE'
]

for col in soc_cols:
    mtraindf[col] = mtraindf[col].fillna(mtraindf[col].median())
```

4. AMT_ANNUITY — super important

```python
mtraindf['AMT_ANNUITY'] = mtraindf['AMT_ANNUITY'].fillna(mtraindf['AMT_ANNUITY'].median())
```

5. EXT_SOURCE_2  

```python
mtraindf['EXT_SOURCE_2'] = mtraindf['EXT_SOURCE_2'].fillna(mtraindf['EXT_SOURCE_2'].median())
```

6. ORGANIZATION_TYPE_ columns → 0 missing 
      ->one-hot encoded columns->
      no missing here.

```python
mtraindf.isnull().sum().sum()

```

DAYS_EMPLOYED = 365243 (The BIGGEST problem in the dataset)

This value means “not employed” (Home Credit used it as a placeholder).

```python
mtraindf['DAYS_EMPLOYED'].replace(365243, np.nan, inplace=True)
```

```python
mtraindf['DAYS_EMPLOYED'] = mtraindf['DAYS_EMPLOYED'].fillna(mtraindf['DAYS_EMPLOYED'].median())
```

 AMT_INCOME_TOTAL (Outlier)

Some applicants have extremely fake values (25 million income ).

 

```python
upper_limit = mtraindf['AMT_INCOME_TOTAL'].quantile(0.99)
mtraindf['AMT_INCOME_TOTAL'] = np.where(
    mtraindf['AMT_INCOME_TOTAL'] > upper_limit,
    upper_limit,
    mtraindf['AMT_INCOME_TOTAL']
)
```

AMT_CREDIT, AMT_ANNUITY, AMT_GOODS_PRICE → some extreme values

```python
for col in ['AMT_CREDIT', 'AMT_ANNUITY', 'AMT_GOODS_PRICE']:
    upper = mtraindf[col].quantile(0.99)
    mtraindf[col] = np.where(mtraindf[col] > upper, upper, mtraindf[col])
```

CNT_FAM_MEMBERS (Weird values like 15, 20)

Family size > 10 is basically impossible and always noise.

```python
mtraindf['CNT_FAM_MEMBERS'] = np.where(
    mtraindf['CNT_FAM_MEMBERS'] > 10,
    10,
    mtraindf['CNT_FAM_MEMBERS']
)
```

REGION_RATING_CLIENT_W_CITY (values 0 → mistake)

If you see 0, replace with most common (mode):

```python
mtraindf['REGION_RATING_CLIENT_W_CITY'] = mtraindf['REGION_RATING_CLIENT_W_CITY'].replace(
    0,
    mtraindf['REGION_RATING_CLIENT_W_CITY'].mode()[0]
)
```

```python
import numpy as np

mtestdf = pd.read_csv('application_test.csv')

cols_to_drop = [
    'COMMONAREA_MEDI','COMMONAREA_MODE','COMMONAREA_AVG',
    'NONLIVINGAPARTMENTS_MEDI','NONLIVINGAPARTMENTS_MODE','NONLIVINGAPARTMENTS_AVG',
    'LIVINGAPARTMENTS_MEDI','LIVINGAPARTMENTS_MODE','LIVINGAPARTMENTS_AVG',
    'FLOORSMIN_MEDI','FLOORSMIN_MODE','FLOORSMIN_AVG',
    'OWN_CAR_AGE',
    'YEARS_BUILD_AVG','YEARS_BUILD_MEDI','YEARS_BUILD_MODE',
    'LANDAREA_MODE','LANDAREA_AVG','LANDAREA_MEDI',
    'BASEMENTAREA_MEDI'
]

cols_to_drop_2 = [
    'BASEMENTAREA_MODE','BASEMENTAREA_AVG',
    'NONLIVINGAREA_MEDI','NONLIVINGAREA_MODE','NONLIVINGAREA_AVG'
]

cols_to_drop_3 = [
    'APARTMENTS_MODE','APARTMENTS_MEDI','APARTMENTS_AVG',
    'ENTRANCES_MODE','ENTRANCES_AVG','ENTRANCES_MEDI',
    'LIVINGAREA_MODE','LIVINGAREA_MEDI','LIVINGAREA_AVG',
    'FLOORSMAX_MEDI','FLOORSMAX_MODE','FLOORSMAX_AVG',
    'YEARS_BEGINEXPLUATATION_MODE','YEARS_BEGINEXPLUATATION_AVG','YEARS_BEGINEXPLUATATION_MEDI',
    'TOTALAREA_MODE'
]

mtestdf = mtestdf.drop(columns=cols_to_drop, errors='ignore')
mtestdf = mtestdf.drop(columns=cols_to_drop_2, errors='ignore')
mtestdf = mtestdf.drop(columns=cols_to_drop_3, errors='ignore')

mtestdf['DAYS_EMPLOYED'].replace(365243, np.nan, inplace=True)
mtestdf['DAYS_EMPLOYED'] = mtestdf['DAYS_EMPLOYED'].fillna(mtraindf['DAYS_EMPLOYED'].median())

bureau_cols = [
    'AMT_REQ_CREDIT_BUREAU_WEEK','AMT_REQ_CREDIT_BUREAU_QRT',
    'AMT_REQ_CREDIT_BUREAU_MON','AMT_REQ_CREDIT_BUREAU_YEAR',
    'AMT_REQ_CREDIT_BUREAU_DAY','AMT_REQ_CREDIT_BUREAU_HOUR'
]

for col in bureau_cols:
    if col in mtestdf.columns:
        mtestdf[col] = mtestdf[col].fillna(0)

mtestdf['EXT_SOURCE_3'] = mtestdf['EXT_SOURCE_3'].fillna(mtraindf['EXT_SOURCE_3'].median())
mtestdf['EXT_SOURCE_2'] = mtestdf['EXT_SOURCE_2'].fillna(mtraindf['EXT_SOURCE_2'].median())
mtestdf['AMT_ANNUITY'] = mtestdf['AMT_ANNUITY'].fillna(mtraindf['AMT_ANNUITY'].median())

soc_cols = [
    'OBS_30_CNT_SOCIAL_CIRCLE','DEF_30_CNT_SOCIAL_CIRCLE',
    'OBS_60_CNT_SOCIAL_CIRCLE','DEF_60_CNT_SOCIAL_CIRCLE'
]

for col in soc_cols:
    if col in mtestdf.columns:
        mtestdf[col] = mtestdf[col].fillna(mtraindf[col].median())

upper_income = mtraindf['AMT_INCOME_TOTAL'].quantile(0.99)
mtestdf['AMT_INCOME_TOTAL'] = np.where(mtestdf['AMT_INCOME_TOTAL'] > upper_income, upper_income, mtestdf['AMT_INCOME_TOTAL'])

for col in ['AMT_CREDIT','AMT_ANNUITY','AMT_GOODS_PRICE']:
    if col in mtestdf.columns:
        upper = mtraindf[col].quantile(0.99)
        mtestdf[col] = np.where(mtestdf[col] > upper, upper, mtestdf[col])

mtestdf = pd.get_dummies(mtestdf, columns=cat_cols, drop_first=True)

mtraindf, mtestdf = mtraindf.align(mtestdf, join='left', axis=1)
mtestdf = mtestdf.fillna(0)

mtestdf.isnull().sum().sum()
```
